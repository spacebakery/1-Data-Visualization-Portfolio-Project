```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# load data
df = pd.read_csv('all_data.csv')
print(df.Country.value_counts())
df.head()
```

```python
# Rename column 'Life expectancy at birth (years)' to 'LEABY'
df.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns", inplace=True)
df.columns
```

```python
sns.set_theme(style="whitegrid")
```

```python
sns.scatterplot(data=df, x='GDP', y='LEABY', hue='Country')
plt.show()
```

```python
plt.figure(figsize=(8,5))
# sns.set_theme(style="whitegrid")


g = sns.relplot(
    data=df,
    x="GDP", y="LEABY",
    hue="Country", size='GDP',
    palette='deep', sizes=(10, 500),
    alpha=0.8
)
# g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
sns.lmplot(data=df, x='GDP', y='LEABY', hue='Country', fit_reg=True, line_kws={'color': 'black', 'alpha':1, 'lw':1}, lowess=True)

plt.show()
```

```python
print(df.Country.unique())

plt.figure(figsize=(10,7))
sns.lineplot(data=df, x='Year', y=df.GDP[df.Country == 'Chile'], marker='s', label='Chile')
sns.lineplot(data=df, x='Year', y=df.GDP[df.Country == 'China'], marker='s', label='China')
sns.lineplot(data=df, x='Year', y=df.GDP[df.Country == 'Germany'], marker='s', label='ChiGermanyle')
sns.lineplot(data=df, x='Year', y=df.GDP[df.Country == 'Mexico'], marker='s', label='ChMexicoile')
sns.lineplot(data=df, x='Year', y=df.GDP[df.Country == 'United States of America'], marker='s', label='United States of America')
sns.lineplot(data=df, x='Year', y=df.GDP[df.Country == 'Zimbabwe'], marker='s', label='Zimbabwe')

plt.title('GDP per Country')
plt.xticks(np.arange(2000, 2016))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(data=df, x='Year', y=df['LEABY'][df.Country == 'Chile'], marker='s', label='Chile')
sns.lineplot(data=df, x='Year', y=df['LEABY'][df.Country == 'China'], marker='s', label='China')
sns.lineplot(data=df, x='Year', y=df['LEABY'][df.Country == 'Germany'], marker='s', label='Germany')
sns.lineplot(data=df, x='Year', y=df['LEABY'][df.Country == 'Mexico'], marker='s', label='Mexico')
sns.lineplot(data=df, x='Year', y=df['LEABY'][df.Country == 'United States of America'], marker='s', label='United States of America')
sns.lineplot(data=df, x='Year', y=df['LEABY'][df.Country == 'Zimbabwe'], marker='s', label='Zimbabwe')

plt.ylabel('Life expectancy at birth (years)')
plt.title('Life expectancy at birth (years) per Country')
plt.xticks(np.arange(2000, 2016))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(df, x='Year', y='GDP', hue='Country')
plt.yscale('log')
plt.xticks(rotation=45, ha='right')
plt.xlabel('')
plt.ylabel('Gross Domestic Product in USD (GDP)')
plt.title('Gross Domestic Product in USD 2000-2015')
plt.legend(loc='upper right', bbox_to_anchor=(1.4, 0.7))
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
sns.barplot(df, x='Year', y='LEABY', hue='Country')
plt.yscale('log')
plt.xticks(rotation=45, ha='right')
plt.xlabel('')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Life expectancy at birth (years) 2000-2015')
plt.legend(loc='upper right', bbox_to_anchor=(1.4, 0.7))
plt.show()
```
